```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from sklearn.neighbors import NearestNeighbors

# Read location data from CSV file
locations = []
with open('C:/Users/vishn/OneDrive/Documents/BTP/FINAL/FINAL_LATITUDE_LONGITUDE.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    for row in reader:
        locations.append([float(row[3]), float(row[4])])

num_all_locations = len(locations)

# Convert locations into a NumPy array
coordinates = np.array(locations)

# Calculate distances between points
distances = []
for i in range(num_all_locations):
    for j in range(i + 1, num_all_locations):
        dist = np.sqrt(np.sum((coordinates[i] - coordinates[j]) ** 2))
        distances.append(dist)

# Calculate semivariance
semivariance = []
lags = np.arange(0, max(distances), 1)  # Define the lag distances

for lag in lags:
    pairs = []
    for i in range(len(distances)):
        if distances[i] >= lag:
            pairs.append(i)

    if len(pairs) > 0:
        semivar = np.var([distances[p] for p in pairs])
        semivariance.append(semivar)
    else:
        semivariance.append(np.nan)

# Plot variogram
plt.plot(lags, semivariance, 'bo-')
plt.xlabel('Lag Distance')
plt.ylabel('Semivariance')
plt.title('Variogram')
plt.grid(True)

# Calculate nearest neighbor connections
nn = NearestNeighbors(n_neighbors=2, metric='haversine')
nn.fit(np.radians(coordinates))
_, indices = nn.kneighbors(coordinates)

# Plot the connections between nodes
for i in range(num_all_locations):
    nearest_neighbor = indices[i][1]
    start_location = locations[i]
    end_location = locations[nearest_neighbor]
    plt.plot([start_location[1], end_location[1]], [start_location[0], end_location[0]], 'r--')

plt.legend(['Variogram', 'Nearest Neighbor Connection'])
plt.show()

# Print which node is nearest to which node
print("Node Connections:")
for i in range(num_all_locations):
    nearest_neighbor = indices[i][1]
    print(f"Node {i}: Nearest Neighbor {nearest_neighbor}")

# Save adjacency matrix to CSV file
adjacency_matrix = np.zeros((num_all_locations, num_all_locations))
for i in range(num_all_locations):
    nearest_neighbor = indices[i][1]  # Index 0 is the location itself, so index 1 is the nearest neighbor
    adjacency_matrix[i][nearest_neighbor] = 1

with open('adjacency_matrix.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(adjacency_matrix)

print("Adjacency matrix saved to adjacency_matrix.csv")
```

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from sklearn.neighbors import NearestNeighbors

# Read location data from CSV file
locations = []
with open('C:/Users/vishn/OneDrive/Documents/BTP/FINAL/FINAL_LATITUDE_LONGITUDE.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    for row in reader:
        locations.append([float(row[3]), float(row[4])])

num_all_locations = len(locations)

# Convert locations into a NumPy array
coordinates = np.array(locations)

# Calculate distances between points
distances = []
for i in range(num_all_locations):
    for j in range(i + 1, num_all_locations):
        dist = np.sqrt(np.sum((coordinates[i] - coordinates[j]) ** 2))
        distances.append(dist)

# Calculate semivariance
semivariance = []
lags = np.arange(0, max(distances), 1)  # Define the lag distances

for lag in lags:
    pairs = []
    for i in range(len(distances)):
        if distances[i] >= lag:
            pairs.append(i)

    if len(pairs) > 0:
        semivar = np.var([distances[p] for p in pairs])
        semivariance.append(semivar)
    else:
        semivariance.append(np.nan)

# Plot variogram
plt.plot(lags, semivariance, 'bo-')
plt.xlabel('Lag Distance')
plt.ylabel('Semivariance')
plt.title('Variogram')
plt.grid(True)

# Calculate nearest neighbor connections for all clusters
nn = NearestNeighbors(n_neighbors=2, metric='haversine')
nn.fit(np.radians(coordinates))
_, indices = nn.kneighbors(coordinates)

# Select the first three clusters
num_clusters = 3
cluster_indices = np.array_split(np.arange(num_all_locations), num_clusters)

# Plot the connections between nodes for the selected clusters
for cluster_index in cluster_indices:
    for i in cluster_index:
        nearest_neighbor = indices[i][1]
        start_location = locations[i]
        end_location = locations[nearest_neighbor]
        plt.plot([start_location[1], end_location[1]], [start_location[0], end_location[0]], 'r--')

plt.legend(['Variogram', 'Nearest Neighbor Connection'])
plt.show()

# Print which node is nearest to which node for the selected clusters
print("Node Connections for the Selected Clusters:")
for cluster_index in cluster_indices:
    for i in cluster_index:
        nearest_neighbor = indices[i][1]
        print(f"Node {i}: Nearest Neighbor {nearest_neighbor}")

# Save adjacency matrix for the selected clusters to CSV file
adjacency_matrix = np.zeros((num_all_locations, num_all_locations))
for cluster_index in cluster_indices:
    for i in cluster_index:
        nearest_neighbor = indices[i][1]  # Index 0 is the location itself, so index 1 is the nearest neighbor
        adjacency_matrix[i][nearest_neighbor] = 1

with open('adjacency_matrix_selected_clusters.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(adjacency_matrix)

print("Adjacency matrix for the selected clusters saved to adjacency_matrix_selected_clusters.csv")
```
